```python
import pandas as pd
import json
```

```python
malicious_file = "../data/raw/malicous_deepset.csv"
df_malicious = pd.read_csv(malicious_file)
```

```python
df_malicious = df_malicious.dropna(subset=["Prompt"])
```

```python
df_malicious_sample = df_malicious.head(50).copy()
```

```python
print(df_malicious_sample.head())
```

```python
df_malicious_sample["id"] = [f"id_1_{i+1}" for i in range(len(df_malicious_sample))]
df_malicious_sample["label"] = "injection"
```

```python
malicious_jsonl = "../data/cleaned/50_malicious_prompts.jsonl"
with open(malicious_jsonl, "w", encoding="utf-8") as f:
    for _, row in df_malicious_sample.iterrows():
        json_line = {
            "id": row["id"],
            "prompt": row["Prompt"],
            "label": row["label"]
        }
        f.write(json.dumps(json_line, ensure_ascii=False) + "\n")

print(f"Saved benign prompts to {malicious_jsonl}")
```
