```python
import pandas as pd
import json
```

```python
benign_file2 = "../data/raw/benign_prompts_chatgpt.csv"
df_benign2 = pd.read_csv(benign_file2)
```

```python
df_benign2 = df_benign2.dropna(subset=["Prompt"])
```

```python
print(df_benign2.head())
```

```python
df_benign2["id"] = [f"id_1_{i+1}" for i in range(len(df_benign2))]
df_benign2["label"] = "benign"
```

```python
df_benign_sample = df_benign2.sample(n=50, random_state=42).copy()
```

```python
benign_jsonl2 = "../data/cleaned/benign_prompts_chatgpt.jsonl"
with open(benign_jsonl2, "w", encoding="utf-8") as f:
    for _, row in df_benign_sample.iterrows():
        json_line = {
            "id": row["id"],
            "prompt": row["Prompt"],
            "label": row["label"]
        }
        f.write(json.dumps(json_line, ensure_ascii=False) + "\n")

print(f"Saved benign prompts to {benign_jsonl2}")
```

```python
import json
import random

input_file = "../data/raw/Prompt_INJECTION_And_Benign_DATASET.jsonl"
output_file = "../data/cleaned/cleanedPrompt_INJECTION_And_Benign_DATASET_sampled_50.jsonl"

cleaned_data = []

# Step 1: Read and clean
with open(input_file, "r", encoding="utf-8") as fin:
    for line in fin:
        obj = json.loads(line)

        # Map label
        if obj["label"] in ("none", "benign"):
            new_label = "benign"
        else:
            continue

        cleaned_data.append({
            "prompt": obj["prompt"],
            "label": new_label
        })

# Step 2: Randomly sample 50
sample_size = min(50, len(cleaned_data))
sampled_data = random.sample(cleaned_data, sample_size)

# Step 3: Assign IDs and enforce key order
with open(output_file, "w", encoding="utf-8") as fout:
    for i, item in enumerate(sampled_data):
        record = {
            "id": f"id_1_{i + 1}",
            "prompt": item["prompt"],
            "label": item["label"]
        }
        fout.write(json.dumps(record, ensure_ascii=False) + "\n")

print(f"Saved {sample_size} samples with ordered fields to {output_file}")
```

```python
file1 = "../data/cleaned/benign_prompts_chatgpt.jsonl"
file2 = "../data/cleaned/cleanedPrompt_INJECTION_And_Benign_DATASET_sampled_50.jsonl"
output_merged = "../data/cleaned/100_benign_prompts.jsonl"
```

```python
with open(output_merged, "w", encoding="utf-8") as outfile:
    for fname in [file1, file2]:
        with open(fname, "r", encoding="utf-8") as infile:
            for line in infile:
                outfile.write(line)
print(f"Merged JSONL file created: {output_merged}")
```
