Preprocessing jailbreak csv

```python
import pandas as pd
import json
```

```python
df = pd.read_csv('../data/raw/SimsonsunJailbreakPrompts.csv')
```

```python
print(df.head())
```

```python
df = df[['Prompt']].dropna()
```

```python
df_sample = df.sample(n=50, random_state=42).copy()
```

```python
df_sample["id"] = [f"id_1_{i+1}" for i in range(len(df_sample))]
df_sample["label"] = "injection" 
df_sample = df_sample[["id", "Prompt", "label"]]
```

```python
output_file = "../data/cleaned/jailbreak_prompts_Simsonsun.jsonl"
with open(output_file, "w", encoding="utf-8") as f:
    for _, row in df_sample.iterrows():
        json_line = {
            "id": row["id"],
            "prompt": row["Prompt"],
            "label": row["label"]
        }
        f.write(json.dumps(json_line, ensure_ascii=False) + "\n")
```

```python
print(f"Random 50 JSONL file created: {output_file}")
```

```python
df2 = pd.read_csv('../data/raw/Bravanskycompact-jailbreaks.csv')
```

```python
df2 = df2.dropna(subset=["attack"])
```

```python
df_sample2 = df2.sample(n=50, random_state=42).copy()
```

```python
df_sample2["id"] = [f"id_2_{i+1}" for i in range(len(df_sample2))]
df_sample2["label"] = "injection"
```

```python
output_file2 = "../data/cleaned/jailbreak_prompts_Bravansky.jsonl"
with open(output_file2, "w", encoding="utf-8") as f:
    for _, row in df_sample2.iterrows():
        json_line = {
            "id": row["id"],
            "prompt": row["attack"],  # lowercase 'prompt'
            "label": row["label"]
        }
        f.write(json.dumps(json_line, ensure_ascii=False) + "\n")
```

```python
file1 = "../data/cleaned/jailbreak_prompts_Simsonsun.jsonl"
file2 = "../data/cleaned/jailbreak_prompts_Bravansky.jsonl"
output_merged = "../data/cleaned/100_jailbreak_prompts.jsonl"
```

```python
with open(output_merged, "w", encoding="utf-8") as outfile:
    for fname in [file1, file2]:
        with open(fname, "r", encoding="utf-8") as infile:
            for line in infile:
                outfile.write(line)
print(f"Merged JSONL file created: {output_merged}")
```
