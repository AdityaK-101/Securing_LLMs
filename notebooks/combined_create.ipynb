```python
import pandas as pd
import json
```

```python
files = [
    "../data/cleaned/50_malicious_prompts.jsonl",
    "../data/cleaned/100_benign_prompts.jsonl",
    "../data/cleaned/100_jailbreak_prompts.jsonl"
]
```

```python
dfs = []
for f in files:
    with open(f, "r", encoding="utf-8") as file:
        data = [json.loads(line) for line in file if line.strip()]
        df = pd.DataFrame(data)
        dfs.append(df)
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
combined_df['id'] = ["id_" + str(i+1) for i in range(len(combined_df))]
```

```python
combined_df.to_csv("../data/cleaned/combined_dataset.csv", index=False, encoding="utf-8")
```

```python
print(f"Combined dataset saved! Total rows: {len(combined_df)}")
```
